
ğŸ“˜ What Is N-HiTS?
N-HiTS (Neural Hierarchical Interpolation for Time Series) is a state-of-the-art neural forecasting model introduced in 2022 by researchers from Amazon AWS Labs. It builds on ideas from the N-BEATS family, but with improved long-horizon forecasting accuracy, computational efficiency, and scalability.

ğŸ§  Paper: "N-HiTS: Neural Hierarchical Interpolation for Time Series Forecasting"

ğŸ› ï¸ Library: Available in PyTorch via Darts

ğŸ§  How N-HiTS Works (Simplified)
1. Stacked Neural Blocks
N-HiTS consists of multiple independent blocks (stacks). Each block:

Processes the time series with a multi-layer perceptron (MLP).

Predicts part of the forecast.

Refines the remaining error (residual learning).

This layered refinement approach enables it to learn both short- and long-term patterns more effectively.

2. Hierarchical Interpolation
Unlike traditional models that operate directly on raw input:

N-HiTS projects past values into interpolated basis representations (low-dimensional space).

These are resampled (hierarchically) to match the forecast horizon.

This makes N-HiTS especially good at:
âœ… Capturing multiple seasonalities
âœ… Handling long forecast windows
âœ… Working with high-frequency or uneven time data

```python
# Step 1: Install required packages (if not already done)
!pip install u8darts[torch] --quiet

# Step 2: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import NHiTSModel
from darts.metrics import mape
from darts.utils.timeseries_generation import datetime_attribute_timeseries

# Step 3: Load your dataset
df = pd.read_csv("/content/monthly-milk-production-pounds.csv")
df.columns = ['timestamp', 'value']
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Step 4: Create Darts TimeSeries object
series = TimeSeries.from_dataframe(df, time_col="timestamp", value_cols="value", freq="MS")

# Step 5: Train/validation split
train, val = series[:-12], series[-12:]

# Step 6: Define and train the NHiTS model
model = NHiTSModel(
    input_chunk_length=36,
    output_chunk_length=12,
    n_epochs=200,
    random_state=42,
    batch_size=32
)
model.fit(train)

# Step 7: Forecast
forecast = model.predict(n=12)

# Step 8: Plot
series[-36:].plot(label="Actual")
forecast.plot(label="NHiTS Forecast")
plt.title("ğŸ“ˆ NHiTSModel Forecast")
plt.grid(True)
plt.legend()
plt.show()

# Step 9: Accuracy
print(f"MAPE (12-month forecast): {mape(val, forecast):.2f}%")

# Step 10: Comparison table
comparison_df = pd.DataFrame({
    'timestamp': val.time_index,
    'actual': val.values().flatten(),
    'forecast': forecast.values().flatten()
})
print("\nComparison Table:")
print(comparison_df)
```

```python
#Export results
comparison_df.to_csv("milk_forecast_nhits.csv", index=False)
```

```python
#To generate true future forecast (e.g., for 1976 and beyond):


# Predict into the future (12 months ahead from the last point in series)
future_forecast = model.predict(n=12)

# Create a DataFrame for future-only prediction
future_df = pd.DataFrame({
    "timestamp": future_forecast.time_index,
    "forecast": future_forecast.values().flatten()
})
print(future_df)
```

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import NHiTSModel
from darts.metrics import mape
from darts.utils.timeseries_generation import datetime_attribute_timeseries

# Load the dataset
df = pd.read_csv("/content/monthly-milk-production-pounds.csv")
df.columns = ["timestamp", "value"]
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Create TimeSeries object with monthly start frequency
series = TimeSeries.from_dataframe(df, time_col="timestamp", value_cols="value", freq="MS")

# Create monthly covariates (e.g. January, February, etc.)
month_cov = datetime_attribute_timeseries(series, attribute="month", one_hot=True)

# Train on full dataset
model = NHiTSModel(
    input_chunk_length=24,
    output_chunk_length=12,
    n_epochs=200,
    random_state=42,
    batch_size=16
)

model.fit(series, past_covariates=month_cov)

# Forecast 12 months into the future (beyond dataset)
forecast = model.predict(n=12, past_covariates=month_cov)

# Build future timestamps manually
future_dates = pd.date_range(start=series.end_time() + pd.DateOffset(months=1), periods=12, freq="MS")

# Build forecast DataFrame
future_df = pd.DataFrame({
    "timestamp": future_dates,
    "forecast": forecast.values().flatten()
})

# Show future predictions
print("ğŸ“… Future Forecast:")
print(future_df)

# Plot actual + future forecast
series.plot(label="Historical")
forecast.plot(label="Future Forecast")
plt.title("ğŸ“ˆ NHiTS Forecast for Milk Production (Future 12 Months)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# ğŸ“š Imports
import pandas as pd
from darts import TimeSeries
from darts.models import NHiTSModel
from darts.metrics import mape
import matplotlib.pyplot as plt
```

```python
# ğŸ“ Load dataset
df = pd.read_csv("/content/avonline_forecast_data.csv")
df['date'] = pd.to_datetime(df['date'])
```

```python
# ğŸ¯ Filter one region
region = 'North'
df_region = df[df['region'] == region].copy()

# ğŸ•’ Create TimeSeries
series = TimeSeries.from_dataframe(df_region, time_col='date', value_cols='new_connections', freq='D')
train, val = series[:-30], series[-30:]
```
